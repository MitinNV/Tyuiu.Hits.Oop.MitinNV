````
using System;

// üîπ –ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å IAnimal
public interface IAnimal
{
    void Speak();
    void Move();
    void Eat(string food);
}

// üîπ –ê–±—Å—Ç—Ä–∞–∫—Ç–Ω—ã–π –±–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å Animal, —Ä–µ–∞–ª–∏–∑—É—é—â–∏–π –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å
public abstract class Animal : IAnimal
{
    private string name;
    private int age;
    private double weight;
    private string species;

    public string Name
    {
        get { return name; }
        set { name = value.Length > 0 ? value : "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ"; }
    }

    public int Age
    {
        get { return age; }
        set { age = value > 0 ? value : 1; }
    }

    public double Weight
    {
        get { return weight; }
        set { weight = value > 0 ? value : 0.1; }
    }

    public string Species
    {
        get { return species; }
        set { species = value; }
    }

    public Animal(string name, int age, double weight, string species)
    {
        Name = name;
        Age = age;
        Weight = weight;
        Species = species;
    }

    public abstract void Speak();
    public abstract void Move();

    public virtual void Eat(string food)
    {
        Console.WriteLine($"{Name} –µ—Å—Ç {food}.");
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"–ñ–∏–≤–æ—Ç–Ω–æ–µ: {Species}, –ò–º—è: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç, –í–µ—Å: {Weight} –∫–≥.");
    }
}

// üîπ –ö–ª–∞—Å—Å Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, double weight, string breed)
        : base(name, age, weight, "–°–æ–±–∞–∫–∞")
    {
        Breed = breed;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} –ª–∞–µ—Ç: –ì–∞–≤-–≥–∞–≤!");
    }

    public override void Move()
    {
        Console.WriteLine($"{Name} –±–µ–∂–∏—Ç –ø–æ –¥–≤–æ—Ä—É.");
    }

    public override void Eat(string food)
    {
        Console.WriteLine($"{Name} —Å —É–¥–æ–≤–æ–ª—å—Å—Ç–≤–∏–µ–º –µ—Å—Ç {food} –∏ –≤–∏–ª—è–µ—Ç —Ö–≤–æ—Å—Ç–æ–º.");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–ü–æ—Ä–æ–¥–∞: {Breed}");
    }
}

// üîπ –ö–ª–∞—Å—Å Cat
public class Cat : Animal
{
    public bool IsDomestic { get; set; }

    public Cat(string name, int age, double weight, bool isDomestic)
        : base(name, age, weight, "–ö–æ—à–∫–∞")
    {
        IsDomestic = isDomestic;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} –º—è—É–∫–∞–µ—Ç: –ú—è—É-–º—è—É!");
    }

    public override void Move()
    {
        Console.WriteLine($"{Name} –≥—Ä–∞—Ü–∏–æ–∑–Ω–æ –∏–¥–µ—Ç –ø–æ –ø–æ–¥–æ–∫–æ–Ω–Ω–∏–∫—É.");
    }

    public override void Eat(string food)
    {
        Console.WriteLine($"{Name} –∞–∫–∫—É—Ä–∞—Ç–Ω–æ –µ—Å—Ç {food}.");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–î–æ–º–∞—à–Ω—è—è: {(IsDomestic ? "–î–∞" : "–ù–µ—Ç")}");
    }
}

// üîπ –ö–ª–∞—Å—Å Bird
public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double weight, double wingspan)
        : base(name, age, weight, "–ü—Ç–∏—Ü–∞")
    {
        WingSpan = wingspan;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} –ø–æ—ë—Ç: –ß–∏—Ä–∏–∫-—á–∏—Ä–∏–∫!");
    }

    public override void Move()
    {
        Console.WriteLine($"{Name} –ª–µ—Ç–∏—Ç –ø–æ –Ω–µ–±—É.");
    }

    public override void Eat(string food)
    {
        Console.WriteLine($"{Name} –∫–ª—é—ë—Ç {food}.");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–†–∞–∑–º–∞—Ö –∫—Ä—ã–ª—å–µ–≤: {WingSpan} –º");
    }
}

// üîπ –ö–ª–∞—Å—Å Fish
public class Fish : Animal
{
    public string WaterType { get; set; } // –ü—Ä–µ—Å–Ω–æ–≤–æ–¥–Ω–∞—è / –ú–æ—Ä—Å–∫–∞—è

    public Fish(string name, int age, double weight, string waterType)
        : base(name, age, weight, "–†—ã–±–∞")
    {
        WaterType = waterType;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} –º–æ–ª—á–∏—Ç, –Ω–æ –ø—É—Å–∫–∞–µ—Ç –ø—É–∑—ã—Ä–∏.");
    }

    public override void Move()
    {
        Console.WriteLine($"{Name} –ø–ª—ã–≤—ë—Ç –≤ –≤–æ–¥–µ.");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–¢–∏–ø –≤–æ–¥—ã: {WaterType}");
    }
}
Dog dog = new Dog("–ë–æ–±–∏–∫", 3, 12.5, "–û–≤—á–∞—Ä–∫–∞");
Cat cat = new Cat("–ú—É—Ä–∫–∞", 2, 4.3, true);
Bird bird = new Bird("–ö–µ—à–∞", 1, 0.8, 0.25);
Fish fish = new Fish("–ù–µ–º–æ", 1, 0.2, "–ú–æ—Ä—Å–∫–∞—è");

Console.WriteLine("--- –°–æ–±–∞–∫–∞ ---");
dog.DisplayInfo();
dog.Speak();
dog.Move();
dog.Eat("–∫–æ—Å—Ç—å");

Console.WriteLine("\n--- –ö–æ—à–∫–∞ ---");
cat.DisplayInfo();
cat.Speak();
cat.Move();
cat.Eat("—Ä—ã–±—É");

Console.WriteLine("\n--- –ü—Ç–∏—Ü–∞ ---");
bird.DisplayInfo();
bird.Speak();
bird.Move();
bird.Eat("–∑–µ—Ä–Ω–æ");

Console.WriteLine("\n--- –†—ã–±–∞ ---");
fish.DisplayInfo();
fish.Speak();
fish.Move();
fish.Eat("–ø–ª–∞–Ω–∫—Ç–æ–Ω");
````
