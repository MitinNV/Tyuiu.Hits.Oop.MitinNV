### –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –∏ –û–±—ã—á–Ω—ã–º–∏ –ê—Ç—Ä–∏–±—É—Ç–∞–º–∏ –∏ –ú–µ—Ç–æ–¥–∞–º–∏ –ö–ª–∞—Å—Å–∞

### –ó–∞–∫–ª—é—á–µ–Ω–∏–µ

–í —ç—Ç–æ–º —Ç–∞—Å–∫–µ –º—ã —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–ª–∏ —Å—Ç—Ä—É–∫—Ç—É—Ä—É –∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –≤ C#, –≤–∫–ª—é—á–∞—è —Å–≤–æ–π—Å—Ç–≤–∞, –≥–µ—Ç—Ç–µ—Ä—ã –∏ —Å–µ—Ç—Ç–µ—Ä—ã, –∞ —Ç–∞–∫–∂–µ —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É —Å—Ç–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –∏ –æ–±—ã—á–Ω—ã–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏ –∏ –º–µ—Ç–æ–¥–∞–º–∏ –∫–ª–∞—Å—Å–∞ –Ω–∞ –ø—Ä–∏–º–µ—Ä–∞—Ö –∫–ª–∞—Å—Å–æ–≤ `Vehicle` –∏ `Car`. –≠—Ç–∏ –∫–æ–Ω—Ü–µ–ø—Ü–∏–∏ —è–≤–ª—è—é—Ç—Å—è –æ—Å–Ω–æ–≤–æ–π –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –æ–±—ä–µ–∫—Ç–Ω–æ-–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–æ–≥—Ä–∞–º–º –Ω–∞ C#.

````
using System;

public class Vehicle
{
    private string _color;
    private int _speed;

    // –°–≤–æ–π—Å—Ç–≤–æ –¥–ª—è —Ü–≤–µ—Ç–∞
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // –°–≤–æ–π—Å—Ç–≤–æ –¥–ª—è —Å–∫–æ—Ä–æ—Å—Ç–∏ —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("–°–∫–æ—Ä–æ—Å—Ç—å –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π!");
        }
    }

    // –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π —Å—á—ë—Ç—á–∏–∫ –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤
    public static int TotalVehicles { get; private set; } = 0;

    // –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä
    public Vehicle()
    {
        TotalVehicles++;
    }

    // –ú–µ—Ç–æ–¥ —É–≤–µ–ª–∏—á–µ–Ω–∏—è —Å–∫–æ—Ä–æ—Å—Ç–∏
    public void Accelerate(int delta)
    {
        Speed += delta;
        Console.WriteLine($"–¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç —É—Å–∫–æ—Ä–∏–ª—Å—è. –¢–µ–∫—É—â–∞—è —Å–∫–æ—Ä–æ—Å—Ç—å: {Speed} –∫–º/—á");
    }

    // –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"–í—Å–µ–≥–æ —Å–æ–∑–¥–∞–Ω–æ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö —Å—Ä–µ–¥—Å—Ç–≤: {TotalVehicles}");
    }
}

public class Car : Vehicle
{
    private int _fuelLevel;

    // –°–≤–æ–π—Å—Ç–≤–æ –¥–ª—è —Ç–æ–ø–ª–∏–≤–∞
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä
    public Car() : base()
    {
        FuelLevel = 100; // –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –±–∞–∫ –ø–æ–ª–Ω—ã–π
    }

    // –ú–µ—Ç–æ–¥ –∑–∞–ø—Ä–∞–≤–∫–∏
    public void Refuel(int amount)
    {
        FuelLevel += amount;
        Console.WriteLine($"–ê–≤—Ç–æ–º–æ–±–∏–ª—å –∑–∞–ø—Ä–∞–≤–ª–µ–Ω. –£—Ä–æ–≤–µ–Ω—å —Ç–æ–ø–ª–∏–≤–∞: {FuelLevel}");
    }
}

// –°–æ–∑–¥–∞–µ–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–µ —Å—Ä–µ–¥—Å—Ç–≤–æ
Vehicle v = new Vehicle();
v.Color = "–ë–µ–ª—ã–π";
v.Speed = 50;
v.Accelerate(20);

// –°–æ–∑–¥–∞–µ–º –∞–≤—Ç–æ–º–æ–±–∏–ª—å
Car c = new Car();
c.Color = "–ö—Ä–∞—Å–Ω—ã–π";
c.Speed = 0;
c.Refuel(20);
c.Accelerate(60);

// –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥
Vehicle.DisplayTotalVehicles();
````

````
using System;

// üîπ –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å Animal
public class Animal
{
    private string _name;
    private int _age;

    // –°–≤–æ–π—Å—Ç–≤–∞ —Å –≥–µ—Ç—Ç–µ—Ä–∞–º–∏/—Å–µ—Ç—Ç–µ—Ä–∞–º–∏
    public string Name
    {
        get { return _name; }
        set 
        {
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                _name = "–ë–µ–∑—ã–º—è–Ω–Ω–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ";
        }
    }

    public int Age
    {
        get { return _age; }
        set
        {
            if (value >= 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("–í–æ–∑—Ä–∞—Å—Ç –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º!");
        }
    }

    // –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π –∞—Ç—Ä–∏–±—É—Ç (—Å—á–∏—Ç–∞–µ—Ç –≤—Å–µ—Ö –∂–∏–≤–æ—Ç–Ω—ã—Ö)
    public static int TotalAnimals { get; private set; } = 0;

    // –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä
    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        TotalAnimals++;
    }

    // –û–±—ã—á–Ω—ã–π –º–µ—Ç–æ–¥
    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} –∏–∑–¥–∞–µ—Ç –∑–≤—É–∫...");
    }

    // –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"–í—Å–µ–≥–æ —Å–æ–∑–¥–∞–Ω–æ –∂–∏–≤–æ—Ç–Ω—ã—Ö: {TotalAnimals}");
    }
}

// üîπ –ü—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–π –∫–ª–∞—Å—Å Dog
public class Dog : Animal
{
    private string _breed;

    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} (–ø–æ—Ä–æ–¥–∞ {Breed}) –ª–∞–µ—Ç: –ì–∞–≤-–≥–∞–≤!");
    }

    public void Guard()
    {
        Console.WriteLine($"{Name} –æ—Ö—Ä–∞–Ω—è–µ—Ç –¥–æ–º!");
    }
}

// üîπ –ü—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–π –∫–ª–∞—Å—Å Cat
public class Cat : Animal
{
    public Cat(string name, int age) : base(name, age) { }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} –º—è—É–∫–∞–µ—Ç: –ú—è—É-–º—è—É!");
    }

    public void SleepOnLap()
    {
        Console.WriteLine($"{Name} —Å–ø–∏—Ç –Ω–∞ –∫–æ–ª–µ–Ω—è—Ö —Ö–æ–∑—è–∏–Ω–∞.");
    }
}

// üîπ –¢–µ—Å—Ç –ø—Ä–æ–≥—Ä–∞–º–º—ã
Dog dog = new Dog("–®–∞—Ä–∏–∫", 3, "–û–≤—á–∞—Ä–∫–∞");
Cat cat = new Cat("–ú—É—Ä–∫–∞", 2);

dog.MakeSound();
dog.Guard();

cat.MakeSound();
cat.SleepOnLap();

Animal.DisplayTotalAnimals();
````
